```python
plt.figure(figsize=(10,6))
fil_data = np.squeeze(fil.data)
plt.imshow(fil_data, aspect='auto')
plt.colorbar()
```

```python
def normalize(data, exclude=0.1):
    mean = np.mean(data)
    std = np.std(data)
    data = np.nan_to_num((data-mean)/std)
    return data

plt.figure(figsize=(10,6))
fil_data = normalize(np.squeeze(fil.data),axis=0)
plt.imshow(fil_data, aspect='auto')
plt.colorbar()
```
